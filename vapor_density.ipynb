```python
import numpy as np
import matplotlib.pyplot as plt

# Given data
temperature = [300, 310, 320, 330, 340, 350, 360, 370, 380, 390, 400]
vapor_density = [0.2384, 0.2398, 0.24185, 0.24305, 0.2432, 0.2589, 0.2478, 0.25205, 0.25345, 0.25755, 0.2719]
liquid_density = [0.68285, 0.6695, 0.65875, 0.647, 0.63295, 0.62295, 0.60935, 0.5997, 0.5776, 0.55735, 0.5388]
vapor_error = [0.0012, 0.0026, 0.0021, 0.0003, 0.0068, 0.01, 0.004, 0.0119, 0.0011, 0.0033, 0.0036]
liquid_error = [0.0045, 0.001, 0.0017, 0.0125, 0.0068, 0.0021, 0.0001, 0.0156, 0.0054, 0.0027, 0.01]

# Simulated X-axis (reduced coordinates) for illustration purposes
X = np.linspace(-1, 1, 100)

# Function to generate density profiles based on temperature and dataset
def interpolate_density(X, density):
    # Simple interpolation based on X position
    return np.interp(X, [-1, 0, 1], [density, density + 0.05, density])

# Plotting vapor densities
for i, T in enumerate(temperature):
    plt.plot(X, interpolate_density(X, vapor_density[i]), label=f'Vapor - {T}K', linestyle='--')

# Plotting liquid densities
for i, T in enumerate(temperature):
    plt.plot(X, interpolate_density(X, liquid_density[i]), label=f'Liquid - {T}K', linestyle='-')

# Customize plot to match the journal's style
plt.xlabel('X (reduced coordinate)')
plt.ylabel(r'$\rho$ (g/cm)')
plt.title('Density Profile of Vapor and Liquid Phases')
plt.legend(title='Phases and Temperature')
plt.grid(True)

# Display plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
data = np.array([
    [300, 0.2384, 0.68285, 0.00119999999999998, 0.00450000000000006],
    [310, 0.2398, 0.6695, 0.00260000000000002, 0.001],
    [320, 0.24185, 0.65875, 0.00210000000000002, 0.00169999999999992],
    [330, 0.24305, 0.647, 0.000299999999999995, 0.0125000000000001],
    [340, 0.2432, 0.63295, 0.0068, 0.0068],
    [350, 0.2589, 0.62295, 0.01, 0.00209999999999999],
    [360, 0.2478, 0.60935, 0.004, 0.0001000000000001],
    [370, 0.25205, 0.5997, 0.0119, 0.0156000000000001],
    [380, 0.25345, 0.5776, 0.00109999999999999, 0.00540000000000007],
    [390, 0.25755, 0.55735, 0.00329999999999997, 0.00269999999999992],
    [400, 0.2719, 0.5388, 0.00359999999999999, 0.0099999999999999]
])

# Extract data
temp = data[:, 0]
vapor_density = data[:, 1]
liquid_density = data[:, 2]
vapor_error = data[:, 3]
liquid_error = data[:, 4]

plt.figure(figsize=(12, 6))

# Both Vapor and Liquid Densities
plt.plot(temp, vapor_density, label='Vapor Density', marker='o', linestyle='-')
plt.plot(temp, liquid_density, label='Liquid Density', marker='x', linestyle='--')

# Error bars for both
plt.errorbar(temp, vapor_density, yerr=vapor_error, fmt='o', capsize=5)
plt.errorbar(temp, liquid_density, yerr=liquid_error, fmt='x', capsize=5)

plt.xlabel('Temperature (K)')
plt.ylabel('Density (g/cm)')
plt.title('Vapor and Liquid Density vs. Temperature')
plt.legend()
plt.grid(True)
plt.show()
```
